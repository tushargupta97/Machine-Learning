## $Vectorized-Operation$

```python
import pandas as pd
import numpy as np
```

```python
# What are vectorized operations
a=np.array([1,2,3,4])
a*4    #-> vectorized operations
```

```python
# problem in vectorized opertions in vanilla python
s=['rat','mat','rat',None]
[i.startswith('c') for i in s]
```

```python
# How pandas solves this issue?
s=pd.Series(['cat','mat','rat',None])
s
# string accessor
s.str.startswith('c')     #->vectorized string
# fast and optimized
```

```python
# import titanic
df=pd.read_csv('titanic.csv')
df
df['Name']
```

```python
# Common Functions
# lower/upper/capitalize/title
print(df['Name'].str.lower())
```

```python
print(df['Name'].str.upper())
```

```python
print(df['Name'].str.capitalize())
```

```python
print(df['Name'].str.title())
```

```python
# len
df['Name'].str.len()
```

```python
#Name having most letters
df['Name'].str.len().max()
```

```python
df['Name'].str.len()==82
```

```python
df['Name'][df['Name'].str.len()==82]
```

```python
# strip -> it removes all the leading and trailing spaces
df['Name'].str.strip()
```

```python
# split -> get
df['Name'].str.split(',')
```

```python
df['lastName']=df['Name'].str.split(',').str.get(0)
df.head()
```

```python
df['Name'].str.split(',').str.get(1).str.strip().str.split(' ', n=1)
```

```python
df['Name'].str.split(',').str.get(1).str.strip().str.split(' ', n=1, expand=True)
```

```python
df[['title','firstname']] = df['Name'].str.split(',').str.get(1).str.strip().str.split(' ', n=1, expand=True)
df.head()
```

```python
df['title'].value_counts()
```

```python
# replace
df['title'] = df['title'].str.replace('Ms.','Miss.')
df['title'] = df['title'].str.replace('Mlle.','Miss.')
df['title'].value_counts()
```

```python
# filtering
# startswith/endswith
df[df['firstname'].str.endswith('A')]
```

```python
df['firstname'].str.isdigit()
```

```python
# isdigit/isalpha...
df[df['firstname'].str.isdigit()]
```

```python
# applying regex
# contains
# search john -> both case
df['firstname'].str.contains('john',case=False) #boolean series
df[df['firstname'].str.contains('john',case=False)]
```

```python
# find lastnames with start and end char vowel
df[df['lastName'].str.contains('^[aeiouAEIOU].+[^aeiouAEIOU]$')]
```

```python
# find lastnames with not start and not end char vowel
df[df['lastName'].str.contains('^[^aeiouAEIOU].+[^aeiouAEIOU]$')]
```
