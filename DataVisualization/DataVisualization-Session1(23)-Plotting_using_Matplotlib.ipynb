# Types of Data
* Numerical Data
* Categorical Data

```python
#import the library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

* Bivariate Analysis
* categorical -> numerical and numerical -> numerical
* Use case - Time series data
* x-axis categorical data and y-axis numerical data

```python
# plotting a simple function
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year,price)
```

```python
# from a pandas dataframe
batsman = pd.read_csv('sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# labels title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Rohit Sharma Vs Virat Kohli career comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F109')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D9')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F109',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D9',linestyle='dashdot',linewidth=4)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F109',linestyle='solid',linewidth=3,marker='D')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D9',linestyle='dashdot',linewidth=4,marker='o',markersize=7)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F109',linestyle='solid',linewidth=3,marker='D',label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D9',linestyle='dashdot',linewidth=4,marker='o',markersize=7,label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()
```

```python
# legend -> location
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F109',linestyle='solid',linewidth=3,marker='D',label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D9',linestyle='dashdot',linewidth=4,marker='o',markersize=7,label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc='upper right')
```

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(0,75000)
plt.xlim(2017,2019)
```

```python
# grid
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.grid()
```

```python
# show
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()
plt.show()   #only showed when we use show()  in other code editors
```

```python
# plt.scatter on pandas data
df = pd.read_csv('batter.csv')
df = df.head(50)
df
```

```python
# size
tips = sns.load_dataset('tips')
tips
```

```python
plt.scatter(tips['total_bill'],tips['tip'])
```

```python
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size'])
```

```python
# slower
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20)  #dots also represents the number of customer ie. size
```

## Bar chart
* Bivariate Analysis
* Numerical vs Categorical
* Use case - Aggregate analysis of groups

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color='black')
```

```python
# horizontal bar chart
plt.barh(colors,children,color='black')
```

```python
# color and label
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
df.shape[0]
```

```python
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
plt.bar(df['batsman'],df['2017'])
```

```python
np.arange(df.shape[0])
```

```python
plt.bar(np.arange(df.shape[0])-0.3,df['2015'],width=0.3,color='red')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.3,color='purple')
plt.bar(np.arange(df.shape[0])+0.3,df['2017'],width=0.3,color='grey')
```

```python
plt.bar(np.arange(df.shape[0])-0.3,df['2015'],width=0.3,color='red')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.3,color='purple')
plt.bar(np.arange(df.shape[0])+0.3,df['2017'],width=0.3,color='grey')
plt.xticks(np.arange(df.shape[0]),df['batsman'])
```

```python
plt.bar(np.arange(df.shape[0])-0.3,df['2015'],width=0.3,color='red')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.3,color='purple')
plt.bar(np.arange(df.shape[0])+0.3,df['2017'],width=0.3,color='grey')

plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

## Pie Chart
* Univariate/Bivariate Analysis
* Categorical vs numerical
* Use case - To find contibution on a standard scale

```python
# simple data
data = [23,45,100,20,49]
plt.pie(data)
plt.show()
```

```python
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)
plt.show()
```

```python
# dataset
df = pd.read_csv('gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
#percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
