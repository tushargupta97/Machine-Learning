## Colored Scatterplot

```python
iris=pd.read_csv('iris.csv')
iris.sample(5)
```

```python
iris['Species']=iris['Species'].replace({'Iris-setosa':0,'Iris-virginica':2,'Iris-versicolor':1})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'])
plt.xlabel('sepal length')
plt.ylabel('Petal length')
plt.show()
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter')
plt.xlabel('sepal length')
plt.ylabel('Petal length')
plt.colorbar()
plt.show()
```

```python
# cmap and alpha
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter',alpha=0.5)
plt.xlabel('sepal length')
plt.ylabel('Petal length')
plt.colorbar()
plt.show()
```

## Subplots

```python
batters=pd.read_csv('batter.csv')
batters.head()
```

```python
plt.figure(figsize=(15,6))
plt.scatter(batters['avg'],batters['strike_rate'])
plt.title('Something')
plt.xlabel('Avg')
plt.ylabel('Strike Rate')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,6))
ax.scatter(batters['avg'],batters['strike_rate'],color='red',marker='+')
ax.set_title('Something')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')
fig.show()
```

```python
fig,ax=plt.subplots(nrows=2, ncols=1)
ax[0].scatter(batters['avg'],batters['strike_rate'],color='purple')
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')

ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
fig,ax=plt.subplots(nrows=2, ncols=1,sharex=True)
ax[0].scatter(batters['avg'],batters['strike_rate'],color='purple')
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')

ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
fig,ax=plt.subplots(nrows=2, ncols=2,figsize=(10,10))
ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='purple')
ax[0,1].scatter(batters['avg'],batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])
fig.show()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,1,1) #211
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2 = fig.add_subplot(2,1,2) #212
ax2.hist(batters['runs'])
```

## 3d Surface plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
xx
```

```python
yy.shape
```

```python
z = (xx**2 + yy**2)
z.shape
```

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')

ax.plot_surface(xx,yy,z)
```

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')

p=ax.plot_surface(xx,yy,z, cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx)**2 + np.cos(yy)**2
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='summer')
fig.colorbar(p)
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx, yy = np.meshgrid(x,y)

z = (xx**2 + yy**2)


fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

## Contour Plots
